```python
schemas = ["bronze", "silver", "gold"]

for schema in schemas:
    tables = spark.sql(f"SHOW TABLES IN workspace.{schema}").collect()

    for row in tables:
        if not row.isTemporary:
            full_name = f"workspace.{schema}.{row.tableName}"
            print(f"Dropping {full_name}")
            spark.sql(f"DROP TABLE {full_name}")
```

```python
import os
import shutil
import re

# path
ARCHIVE_DIR = "/Volumes/workspace/raw_data/archives"
RAW_DATA_DIR = "/Volumes/workspace/raw_data/raw_data"

for filename in os.listdir(ARCHIVE_DIR):
    file_path = os.path.join(ARCHIVE_DIR, filename)

    if not os.path.isfile(file_path):
        continue

    # Gestion du catalogue
    if filename == "catalogue.csv":
        dest_path = os.path.join(RAW_DATA_DIR, "catalogue.csv")
        print(f"Déplacement catalogue: {filename} -> {dest_path}")
        shutil.move(file_path, dest_path)
        continue

    # Gestion des ventes
    match = re.match(r"(\w+)_sales_.*\.csv", filename)
    if match:
        ville = match.group(1).lower()
        boutique_dir = os.path.join(RAW_DATA_DIR, f"boutique_{ville}")

        # Création du dossier si nécessaire
        os.makedirs(boutique_dir, exist_ok=True)

        dest_path = os.path.join(boutique_dir, filename)
        print(f"Déplacement ventes: {filename} -> {dest_path}")
        shutil.move(file_path, dest_path)
    else:
        print(f"Erreur avec le fichier: {filename}")
```
